<h1> Remove single value columns </h1>

```python
def distinct_val_of_columns(dataset):
    columns = dataset.columns
    num_val = []
    for column in columns:
        num_val.append(len(dataset[column].unique()))

    col_val_df = pd.DataFrame({'Column': columns, 'Distinct Value': num_val})
    single_val_cols = col_val_df[col_val_df['Distinct Value'] == 1]
    return col_val_df, single_val_cols
```

```python
col_val_df, single_val_cols = distinct_val_of_columns(df)

print(f"{len(single_val_cols)} single value columns")
print(single_val_cols['Column'].values)
```

```python
dataset_preprocessing = df.drop(columns= single_val_cols['Column'].values)
```

```python
dataset_preprocessing
```

<h1> Remove Non-Numeric Value Columns </h1>

```python
numberic_columns = list(dataset_preprocessing.select_dtypes(include=['int',"float"]).columns)

print(f" - Num. numberic columns: {len(numberic_columns)}")
print(f" - Numberic columns: {numberic_columns}")
```

```python
non_numberic_columns = []
for column in dataset_preprocessing.columns:
    if column not in numberic_columns and column != 'Label':
        non_numberic_columns.append(column)

print(f" - Num. non-numberic columns: {len(non_numberic_columns)}")
print(f" - Non-numberic columns: {non_numberic_columns}")
```

```python
# Drop non-numberic columns
dataset_preprocessing = dataset_preprocessing.drop(columns= non_numberic_columns)
```
